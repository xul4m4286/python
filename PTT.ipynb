```python
import requests
from bs4 import BeautifulSoup
```

```python
def getdata(url):
    headers=headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36","cookie":"over18=1"}
    res=requests.get(url,headers=headers)
    soup=BeautifulSoup(res.text)
    results=soup.select('div.title')
    for titles in results:
        titles_a=titles.select_one("a")
        title=titles.text
        if titles_a:
            print(title,'https://www.ptt.cc/'+titles_a.get('href'))
    btn=soup.select('div.btn-group >a')
    nextpage=btn[3].get('href')
    return nextpage
```

```python
url='https://www.ptt.cc/bbs/Gossiping/index.html'
page=0
while page < 4:
    url='https://www.ptt.cc'+getdata(url)
    page+=1
```
